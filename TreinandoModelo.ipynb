```python
import pandas as pd 

dados = pd.read_csv('/content/dados_proficoes_resumo.csv')
```

```python
!pip install pycaret
```

```python
from pycaret.regression import setup, create_model, tune_model, evaluate_model, finalize_model, save_model

modelagem = setup(data = dados,   
                 target = 'Salário',
                  ignore_low_variance = True,
                 combine_rare_levels = True )
```

```python
modelo = create_model('rf')
```

```python
modelo_tunado  = tune_model( modelo, optimize = 'RMSE', n_iter = 10 )
```

```python
evaluate_model(modelo_tunado)
```

```python
final_model = finalize_model(modelo_tunado)
```

```python
save_model(final_model, 'modelo_previsao_salarios')
```
