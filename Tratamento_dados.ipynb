### Tratamento dos dados

```python
 #Profições => padronizar e reduzir a qtde de classes
    
df["('D6', 'anonymized_role')"].unique() #Listar distinct
```

```python
proficao_map = {
                                            'Outras' : 'Outras', 
                    'Data Analyst/Analista de Dados' : 'Analista de Dados',
              'Business Intelligence/Analista de BI' : 'Analista de BI', 
           'Desenvolvedor ou Engenheiro de Software' : 'Desenvolvedor/Engenheiro de Software',
                 'Data Scientist/Cientista de Dados' : 'Cientista de Dados',
               'Analista de Inteligência de Mercado' : 'Analista de BI', 
                                        'Engenheiro' : 'Outras',
             'Business Analyst/Analista de Negócios' : 'Analista de BI',
                 'Data Engineer/Engenheiro de Dados' : 'Engenheiro de Dados',
               'DBA/Administrador de Banco de Dados' : 'Engenheiro de Dados',
                             'Analista de Marketing' : 'Outras',
                                       'Estatístico' : 'Cientista de Dados',
                    'Engenheiro de Machine Learning' : 'Cientista de Dados',
                                         'Economista': 'Outras'
    
}
```

```python
#Tamanho da empresa

df["('P12', 'workers_number')"].unique()
```

```python
tamanho_empresa_map = {
                           'de 1 a 5' : 'Pequena',
                          'de 6 a 10' : 'Pequena',
                         'de 11 a 50' : 'Pequena',
                        'de 51 a 100' : 'Pequena',
                       'de 101 a 500' : 'Media',  
                       'de 501 a 1000': 'Media',
                      'Acima de 3000' : 'Grande', 
                     'de 1001 a 3000' : 'Grande'
}
```

```python
#Salários => transforma o centro de cada categoria em valores numericos (variaveis continuas) 

df["('P16', 'salary_range')"].unique()
```

```python
salarios_map = {
                 'Menos de R$ 1.000/mês' : 1000,
        'de R$ 1.001/mês a R$ 2.000/mês' : 1500,
        'de R$ 2.001/mês a R$ 3000/mês'  : 2500,
        'de R$ 3.001/mês a R$ 4.000/mês' : 3500,
        'de R$ 4.001/mês a R$ 6.000/mês' : 5000,
        'de R$ 6.001/mês a R$ 8.000/mês' : 7000,
        'de R$ 8.001/mês a R$ 12.000/mês': 10000,
      'de R$ 12.001/mês a R$ 16.000/mês' : 14000,
      'de R$ 16.001/mês a R$ 20.000/mês' : 18000,
      'de R$ 20.001/mês a R$ 25.000/mês' : 22500,
                'Acima de R$ 25.001/mês' : 25000
}
```

```python
#Tipo de trabalho => situação

df["('P10', 'job_situation')"].unique()
```

```python
def tipo_trabalho (x) :
    if x in ['Empregado (CTL)', 'Empreendedor ou Empregado (CNPJ)', 'Estagiário']:
        return x
    else:
        return 'Outros'
```

```python
# Idades => reduzir idades em intervalos de corte (classificação)

cortes = [0, 24, 30, 40, 100]
nomes = ['[18,24]','[25,30]','[31,40]','[41,50]']
```

```python
#Montando o DataSet

df['Salário'] = df["('P16', 'salary_range')"].map(salarios_map)
df['Idade'] = pd.cut(df["('P1', 'age')"], bins=cortes, labels=nomes)
df['Profissão'] = df["('D6', 'anonymized_role')"].map(proficao_map)
df['Tamanho da Empresa'] = df["('P12', 'workers_number')"].map(tamanho_empresa_map)
df['Cargo de Gestão'] = df["('P13', 'manager')"].map({0: 'não', 1: 'sim'})
df['Experiência em DS'] = df["('P17', 'time_experience_data_science')"]
df['Tipo de Trabalho'] = df["('P10', 'job_situation')"].apply(tipo_trabalho)
df['Escolaridade'] = df["('P8', 'degreee_level')"]
df['Área de Formação'] = df["('D3', 'anonymized_degree_area')"]
df['Setor de Mercado'] = df["('D4', 'anonymized_market_sector')"]
df['Brasil'] = df["('P3', 'living_in_brasil')"]
df['Estado'] = df["('P5', 'living_state')"]
df['Linguagem Python'] = df["('P21', 'python')"]
df['Linguagem R'] = df["('P21', 'r')"]
df['Linguagem SQL'] = df["('P21', 'sql_')"]

df.head()
```

```python
colunas_uteis = df.columns[-15:]
dados = df[colunas_uteis]

display(dados.head(3))

print(dados.info())
```

```python
# Salvar DataSet resumido => ARQ.CSV

dados.to_csv('dados_proficoes_resumo.csv', index = False)
```




# Usando PYCARET que possibilita a automatização em aplicar modelos de machine learning - nesse caso usaremos o Random Forest

### Apartir desse ponto as execuções seram realizadas no google colab por razões de incompatibilidade do pycaret

from pycaret.regression import setup, create_model, tune_model, evaluate_model, finalize_model, save_model

#Função setup vai efetuar todo pré processamento como tratamento dos dados e dados nulos

modelagem = setup(data = dados,   
                 target = 'Salário',  # Váriavel alvo!
                  ignore_low_variance = True,   # Ignorar variaveis com baixa variancia 
                 combine_rare_levels = True )   

#Função create_model vai criar o modelo Random Forest

modelo = create_model('rf')  #rt referente ao algoritmo do random forest

#Função tune_model melhora a precisão do modelo buscando quais combinações de parametros são mais adequados para esse conjunto de dados

modelo_tunado  = tune_model( modelo, optimize = 'RMSE', n_iter = 10 ) # Grid de 10 possibilidades RMSE

#Função evaluete_model nos mostra gráficos e informações sobre o modelo

evaluate_model(modelo_tunado)

#Finaliza o modelo tunado treinando ele em toda a base de dados

modelo_final = finalize_model(modelo_tunado)

### Salvando o modelo

save_model('modelo_para_previsao_de_salario')
